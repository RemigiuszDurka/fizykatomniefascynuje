# Rzut ukośny - wprowadzenie



## Definicja

**Rzut ukośny** to rodzaj ruchu ciała w polu grawitacyjnym Ziemi, w którym ciało porusza się po torze parabolicznym, rzucone pod pewnym kątem α do poziomu, z określoną prędkością początkową \(v_0\).

## Składowe ruchu

Ruch ten rozkładany jest na dwa niezależne ruchy:

- **Ruch jednostajny prostoliniowy** w poziomie (oś OX)
- **Ruch jednostajnie opóźniony** w pionie (oś OY), pod wpływem przyspieszenia ziemskiego \(g\)

## Równania ruchu

Dla prędkości początkowej \(v_0\) i kąta α:

- Składowe prędkości początkowej:

$$
v_{0x} = v_0 \cos(\alpha), \quad v_{0y} = v_0 \sin(\alpha)
$$

- Równania położenia w czasie:

$$
x(t) = v_0 \cos(\alpha) \cdot t
$$

$$
y(t) = v_0 \sin(\alpha) \cdot t - \frac{1}{2} g t^2
$$

- Równanie toru:

$$
y(x) = x \cdot \tan(\alpha) - \frac{g x^2}{2 v_0^2 \cos^2(\alpha)}
$$

## Wielkości charakterystyczne

- **Zasięg (R):**

$$
R = \frac{v_0^2 \sin(2\alpha)}{g}
$$

- **Maksymalna wysokość (H):**

$$
H = \frac{v_0^2 \sin^2(\alpha)}{2g}
$$

- **Czas lotu (T):**

$$
T = \frac{2 v_0 \sin(\alpha)}{g}
$$

## Przypadek rzutu z wysokości

Gdy ciało zostaje rzucone ukośnie z wysokości \(h\) nad poziomem odniesienia, modyfikujemy równanie położenia w pionie:

$$
y(t) = h + v_0 \sin(\alpha) \cdot t - \frac{1}{2} g t^2
$$

Czas lotu wyznaczany jest z równania:

$$
h + v_0 \sin(\alpha) \cdot t - \frac{1}{2} g t^2 = 0
$$

To równanie kwadratowe można rozwiązać względem \(t\), aby znaleźć czas kontaktu z podłożem.

Zasięg poziomy w tym przypadku:

$$
R = v_0 \cos(\alpha) \cdot T
$$

## Uwagi

- Maksymalny zasięg osiągany jest dla kąta α = 45°
- Tor ruchu ma kształt paraboli skierowanej w dół



# Rzut ukośny z oporem powietrza

Rzut ukośny z oporem powietrza opisuje się za pomocą równań różniczkowych, uwzględniając siłę aerodynamiczną działającą przeciwnie do kierunku ruchu:

$$ \mathbf{F}_{op} = - k \mathbf{v} $$

gdzie:
- $k$ – współczynnik oporu powietrza,
- $\mathbf{v}$ – wektor prędkości.

Układ współrzędnych:
- oś $x$ – pozioma,
- oś $y$ – pionowa.

## Równania ruchu
Z drugiej zasady dynamiki Newtona:

$$ m \frac{d\mathbf{v}}{dt} = \mathbf{F}_{g} + \mathbf{F}_{op} $$

Siły działające na ciało:
- Siła grawitacji: $$ \mathbf{F}_g = (0, -mg) $$
- Siła oporu powietrza: $$ F_{op,x} = - k v_x $$ $$ F_{op,y} = - k v_y $$

Rozkładając na składowe:

$$ \frac{dv_x}{dt} = -\frac{k}{m} v_x $$

$$ \frac{dv_y}{dt} = -g - \frac{k}{m} v_y $$

## Rozwiązania równań
### Składowa pozioma
Rozwiązując równanie różniczkowe:

$$ v_x = v_0 \cos\alpha \cdot e^{-\frac{k}{m} t} $$

Położenie:

$$ x(t) = v_0 \cos\alpha \cdot \frac{m}{k} \left( 1 - e^{-\frac{k}{m} t} \right) $$

### Składowa pionowa
Rozwiązując równanie:

$$ v_y = \left( v_0 \sin\alpha + \frac{mg}{k} \right) e^{-\frac{k}{m} t} - \frac{mg}{k} $$

Położenie:

$$ y(t) = h + \frac{m}{k} \left( v_0 \sin\alpha + \frac{mg}{k} \right) \left( 1 - e^{-\frac{k}{m} t} \right) - \frac{mg}{k} t $$

## Czas lotu i zasięg
Czas lotu $T$ wyznaczamy z równania $y(T) = 0$:

$$ h + \frac{m}{k} \left( v_0 \sin\alpha + \frac{mg}{k} \right) \left( 1 - e^{-\frac{k}{m} T} \right) - \frac{mg}{k} T = 0 $$

Zasięg rzutu:

$$ R = x(T) = v_0 \cos\alpha \cdot \frac{m}{k} \left( 1 - e^{-\frac{k}{m} T} \right) $$

## Wnioski
- Opór powietrza powoduje tłumienie składowych prędkości.
- Maksymalna wysokość i zasięg są mniejsze niż w przypadku rzutu bez oporu.
- Czas lotu wymaga numerycznego rozwiązania równania transcendentalnego.


### Porównanie rzutów ukośnych – bez oporu i z oporem powietrza (liniowym)

| **Cecha / Składnik**                 | **Bez oporu powietrza**                                   | **Z oporem powietrza (liniowym)**                                                                                   |
|--------------------------------------|------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------------------|
| **Siła działająca na ciało**        | $$ \vec{F} = m\vec{g} $$                                   | $$ \vec{F} = m\vec{g} - k\vec{v} $$                                                                                   |
| **Równania ruchu (różniczkowe)**    | $$ \frac{d\vec{v}}{dt} = \vec{g} $$                        | $$ \frac{d\vec{v}}{dt} = \vec{g} - \frac{k}{m}\vec{v} $$                                                              |
| **Prędkość w poziomie**             | $$ v_x(t) = v_0 \cos \theta $$ (stała)                    | $$ v_x(t) = v_{x0} \, e^{- \frac{k}{m} t} $$                                                                          |
| **Prędkość w pionie**               | $$ v_y(t) = v_0 \sin \theta - g t $$                      | $$ v_y(t) = \left( v_{y0} + \frac{mg}{k} \right) e^{- \frac{k}{m} t} - \frac{mg}{k} $$                                |
| **Tor – pozycja x(t)**              | $$ x(t) = v_{x0} \, t $$                                  | $$ x(t) = \frac{m v_{x0}}{k} \left( 1 - e^{- \frac{k}{m} t} \right) $$                                                |
| **Tor – pozycja y(t)**                   | $$ y(t) = v_{y0} \, t - \frac{1}{2} g t^2 $$              | $$ y(t) = \frac{m}{k} \left( v_{y0} + \frac{mg}{k} \right) \left( 1 - e^{- \frac{k}{m} t} \right) - \frac{mg}{k} t $$ |
| **Zasięg**                          | $$ R = \frac{v_0^2 \sin 2\theta}{g} $$                    | Mniejszy, zależny od $k$ – brak prostego wzoru                                                                   |
| **Czas lotu**                       | $$ T = \frac{2 v_0 \sin \theta}{g} $$                     | Mniejszy, zależny od $k$                                                                                          |
| **Tor lotu**                        | Parabola                                                  | Spłaszczona krzywa – nie jest parabolą                                                                               |

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Stałe
g = 9.81  # Przyspieszenie grawitacyjne (m/s²)
v0 = 20   # Prędkość początkowa (m/s)
angle = 45  # Kąt wyrzutu (stopnie)
dt = 0.02  # Krok czasowy (s)
k_m1 = 0.1  # Współczynnik oporu powietrza dla średniego oporu (1/s)
k_m2 = 0.9  # Współczynnik oporu powietrza dla dużego oporu (1/s)

# Składowe prędkości początkowej
v0x = v0 * np.cos(np.radians(angle))
v0y = v0 * np.sin(np.radians(angle))

# Funkcja rozwiązująca równania ruchu numerycznie
def simulate_projectile(v0x, v0y, k_m):
    t, x, y = [0], [0], [0]
    vx, vy = v0x, v0y

    while y[-1] >= 0:
        ax = -k_m * vx
        ay = -g - k_m * vy

        vx += ax * dt
        vy += ay * dt

        x.append(x[-1] + vx * dt)
        y.append(y[-1] + vy * dt)
        t.append(t[-1] + dt)

    return np.array(x), np.array(y)

# Symulacja dla trzech przypadków
x_no_drag, y_no_drag = simulate_projectile(v0x, v0y, 0)  # Bez oporu
x_drag1, y_drag1 = simulate_projectile(v0x, v0y, k_m1)   # Z oporem 0.1
x_drag2, y_drag2 = simulate_projectile(v0x, v0y, k_m2)   # Z oporem 0.9

# Ustawienia wykresu
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, max(x_no_drag.max(), x_drag1.max(), x_drag2.max()) * 1.1)
ax.set_ylim(0, max(y_no_drag.max(), y_drag1.max(), y_drag2.max()) * 1.1)
ax.set_xlabel("Odległość pozioma x (m)")
ax.set_ylabel("Wysokość y (m)")
ax.set_title("Rzut ukośny z różnym oporem powietrza")
ax.grid()

# Obiekty animacji - tylko aktualne pozycje (bez śladu)
ln_no_drag, = ax.plot([], [], 'bo', markersize=8, label="Bez oporu powietrza")
ln_drag1, = ax.plot([], [], 'ro', markersize=8, label="Z oporem (k/m=0.1)")
ln_drag2, = ax.plot([], [], 'go', markersize=8, label="Z oporem (k/m=0.9)")

traj_no_drag, = ax.plot(x_no_drag, y_no_drag, 'b--', alpha=0.5)
traj_drag1, = ax.plot(x_drag1, y_drag1, 'r--', alpha=0.5)
traj_drag2, = ax.plot(x_drag2, y_drag2, 'g--', alpha=0.5)

ax.legend()

# Funkcja inicjalizująca
def init():
    ln_no_drag.set_data([], [])
    ln_drag1.set_data([], [])
    ln_drag2.set_data([], [])
    return ln_no_drag, ln_drag1, ln_drag2

# Funkcja aktualizująca animację - pokazuje tylko aktualną pozycję
def update(frame):
    current_time = frame * dt
    if frame < len(x_no_drag):
        ln_no_drag.set_data([x_no_drag[frame]], [y_no_drag[frame]])
    if frame < len(x_drag1):
        ln_drag1.set_data([x_drag1[frame]], [y_drag1[frame]])
    if frame < len(x_drag2):
        ln_drag2.set_data([x_drag2[frame]], [y_drag2[frame]])
    ax.set_title(f"Rzut ukośny z różnym oporem powietrza (t = {current_time:.2f} s)")
    return ln_no_drag, ln_drag1, ln_drag2

# Tworzenie animacji
frames = max(len(x_no_drag), len(x_drag1), len(x_drag2))
ani = FuncAnimation(fig, update, frames=frames, init_func=init, blit=False, interval=20)

# Wyświetlenie animacji w Colabie
plt.close()
HTML(ani.to_html5_video())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Stałe
v0 = 9    # Prędkość początkowa (m/s)
angle = 45  # Kąt wyrzutu (stopnie)
dt = 0.02  # Krok czasowy (s)
g_earth = 9.81  # Przyspieszenie grawitacyjne na Ziemi (m/s²)
g_mars = 3.72   # Przyspieszenie grawitacyjne na Marsie (m/s²)
g_moon = 1.62   # Przyspieszenie grawitacyjne na Księżycu (m/s²)

# Składowe prędkości początkowej
v0x = v0 * np.cos(np.radians(angle))
v0y = v0 * np.sin(np.radians(angle))

# Obliczenie czasów lotu
T_earth = 2 * v0 * np.sin(np.radians(angle)) / g_earth  # ≈ 1.30 s
T_mars = 2 * v0 * np.sin(np.radians(angle)) / g_mars    # ≈ 3.42 s
T_moon = 2 * v0 * np.sin(np.radians(angle)) / g_moon    # ≈ 7.85 s

# Funkcja rozwiązująca równania ruchu numerycznie (bez oporu powietrza)
def simulate_projectile(v0x, v0y, g):
    t, x, y = [0], [0], [0]
    vx, vy = v0x, v0y

    while y[-1] >= 0:
        ax = 0  # Brak oporu w poziomie
        ay = -g  # Grawitacja w pionie

        vx += ax * dt
        vy += ay * dt

        x.append(x[-1] + vx * dt)
        y.append(y[-1] + vy * dt)
        t.append(t[-1] + dt)

    return np.array(x), np.array(y)

# Symulacja dla trzech przypadków
x_earth, y_earth = simulate_projectile(v0x, v0y, g_earth)  # Ziemia
x_mars, y_mars = simulate_projectile(v0x, v0y, g_mars)    # Mars
x_moon, y_moon = simulate_projectile(v0x, v0y, g_moon)    # Księżyc

# Ustawienia wykresu
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, max(x_earth.max(), x_mars.max(), x_moon.max()) * 1.1)
ax.set_ylim(0, max(y_earth.max(), y_mars.max(), y_moon.max()) * 1.1)
ax.set_xlabel("Odległość pozioma x (m)")
ax.set_ylabel("Wysokość y (m)")
ax.set_title("Rzut ukośny: Ziemia, Mars, Księżyc (v0 = 9 m/s, 45°)")
ax.grid()

# Obiekty animacji - tylko pojedyncze punkty
ln_earth, = ax.plot([], [], 'bo', markersize=8, label=f"Ziemia (g = {g_earth} m/s²)")
ln_mars, = ax.plot([], [], 'ro', markersize=8, label=f"Mars (g = {g_mars} m/s²)")
ln_moon, = ax.plot([], [], 'ko', markersize=8, label=f"Księżyc (g = {g_moon} m/s²)")

ax.legend()

# Linie torów lotu
traj_earth, = ax.plot(x_earth, y_earth, 'b--', alpha=0.5)
traj_mars, = ax.plot(x_mars, y_mars, 'r--', alpha=0.5)
traj_moon, = ax.plot(x_moon, y_moon, 'k--', alpha=0.5)

# Funkcja inicjalizująca
def init():
    ln_earth.set_data([], [])
    ln_mars.set_data([], [])
    ln_moon.set_data([], [])
    return ln_earth, ln_mars, ln_moon

# Funkcja aktualizująca animację - pokazuje tylko aktualną pozycję
def update(frame):
    current_time = frame * dt
    if frame < len(x_earth):
        ln_earth.set_data([x_earth[frame]], [y_earth[frame]])  # Tylko aktualna pozycja
    if frame < len(x_mars):
        ln_mars.set_data([x_mars[frame]], [y_mars[frame]])  # Tylko aktualna pozycja
    if frame < len(x_moon):
        ln_moon.set_data([x_moon[frame]], [y_moon[frame]])  # Tylko aktualna pozycja
    ax.set_title(f"Rzut ukośny: Ziemia, Mars, Księżyc (t = {current_time:.2f} s)")
    return ln_earth, ln_mars, ln_moon

# Tworzenie animacji
frames = int(T_moon / dt) + 1  # Najdłuższy czas (Księżyc)
interval = dt * 1000  # 20 ms między klatkami
ani = FuncAnimation(fig, update, frames=frames, init_func=init, blit=False, interval=interval)

# Wyświetlenie animacji
plt.close()
HTML(ani.to_html5_video())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Stałe
v0 = 9    # Prędkość początkowa (m/s)
angle = 45  # Kąt wyrzutu (stopnie)
dt = 0.02  # Krok czasowy (s)
g_earth = 9.81  # Przyspieszenie grawitacyjne na Ziemi (m/s²)
g_mars = 3.72   # Przyspieszenie grawitacyjne na Marsie (m/s²)
g_moon = 1.62   # Przyspieszenie grawitacyjne na Księżycu (m/s²)
g_neptune = 11.15  # Przyspieszenie grawitacyjne na Neptunie (m/s²)
g_jupiter = 24.79  # Przyspieszenie grawitacyjne na Jowiszu (m/s²)

# Składowe prędkości początkowej
v0x = v0 * np.cos(np.radians(angle))
v0y = v0 * np.sin(np.radians(angle))

# Obliczenie czasów lotu
T_earth = 2 * v0 * np.sin(np.radians(angle)) / g_earth  # ≈ 1.30 s
T_mars = 2 * v0 * np.sin(np.radians(angle)) / g_mars    # ≈ 3.42 s
T_moon = 2 * v0 * np.sin(np.radians(angle)) / g_moon    # ≈ 7.85 s
T_neptune = 2 * v0 * np.sin(np.radians(angle)) / g_neptune  # ≈ 1.15 s
T_jupiter = 2 * v0 * np.sin(np.radians(angle)) / g_jupiter  # ≈ 0.52 s

# Funkcja rozwiązująca równania ruchu numerycznie (bez oporu powietrza)
def simulate_projectile(v0x, v0y, g):
    t, x, y = [0], [0], [0]
    vx, vy = v0x, v0y

    while y[-1] >= 0:
        ax = 0  # Brak oporu w poziomie
        ay = -g  # Grawitacja w pionie

        vx += ax * dt
        vy += ay * dt

        x.append(x[-1] + vx * dt)
        y.append(y[-1] + vy * dt)
        t.append(t[-1] + dt)

    return np.array(x), np.array(y)

# Symulacja dla pięciu przypadków
x_earth, y_earth = simulate_projectile(v0x, v0y, g_earth)  # Ziemia
x_mars, y_mars = simulate_projectile(v0x, v0y, g_mars)    # Mars
x_moon, y_moon = simulate_projectile(v0x, v0y, g_moon)    # Księżyc
x_neptune, y_neptune = simulate_projectile(v0x, v0y, g_neptune)  # Neptun
x_jupiter, y_jupiter = simulate_projectile(v0x, v0y, g_jupiter)  # Jowisz

# Ustawienia wykresu
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_xlim(0, max(x_earth.max(), x_mars.max(), x_moon.max(), x_neptune.max(), x_jupiter.max()) * 1.1)
ax.set_ylim(0, max(y_earth.max(), y_mars.max(), y_moon.max(), y_neptune.max(), y_jupiter.max()) * 1.1)
ax.set_xlabel("Odległość pozioma x (m)")
ax.set_ylabel("Wysokość y (m)")
ax.set_title("Rzut ukośny: Ziemia, Mars, Księżyc, Neptun, Jowisz (v0 = 9 m/s, 45°)")
ax.grid()

# Obiekty animacji - tylko punkty (bez śladu)
ln_earth, = ax.plot([], [], 'bo', label=f"Ziemia (g = {g_earth} m/s²)")
ln_mars, = ax.plot([], [], 'ro', label=f"Mars (g = {g_mars} m/s²)")
ln_moon, = ax.plot([], [], 'ko', label=f"Księżyc (g = {g_moon} m/s²)")
ln_neptune, = ax.plot([], [], 'go', label=f"Neptun (g = {g_neptune} m/s²)")  # Zielony
ln_jupiter, = ax.plot([], [], 'mo', label=f"Jowisz (g = {g_jupiter} m/s²)")  # Magenta

# Linie torów lotu
traj_earth, = ax.plot(x_earth, y_earth, 'b--', alpha=0.5)
traj_mars, = ax.plot(x_mars, y_mars, 'r--', alpha=0.5)
traj_moon, = ax.plot(x_moon, y_moon, 'k--', alpha=0.5)
traj_neptune, = ax.plot(x_neptune, y_neptune, 'g--', alpha=0.5)
traj_jupiter, = ax.plot(x_jupiter, y_jupiter, 'm--', alpha=0.5)

ax.legend()

# Funkcja inicjalizująca
def init():
    ln_earth.set_data([], [])
    ln_mars.set_data([], [])
    ln_moon.set_data([], [])
    ln_neptune.set_data([], [])
    ln_jupiter.set_data([], [])
    return ln_earth, ln_mars, ln_moon, ln_neptune, ln_jupiter

# Funkcja aktualizująca animację z wyświetlaniem czasu
def update(frame):
    current_time = frame * dt
    # Ustawiamy tylko aktualną pozycję (ostatni punkt), nie całą historię
    if frame < len(x_earth):
        ln_earth.set_data([x_earth[frame]], [y_earth[frame]])
    if frame < len(x_mars):
        ln_mars.set_data([x_mars[frame]], [y_mars[frame]])
    if frame < len(x_moon):
        ln_moon.set_data([x_moon[frame]], [y_moon[frame]])
    if frame < len(x_neptune):
        ln_neptune.set_data([x_neptune[frame]], [y_neptune[frame]])
    if frame < len(x_jupiter):
        ln_jupiter.set_data([x_jupiter[frame]], [y_jupiter[frame]])
    ax.set_title(f"Rzut ukośny: Ziemia, Mars, Księżyc, Neptun, Jowisz (t = {current_time:.2f} s)")
    return ln_earth, ln_mars, ln_moon, ln_neptune, ln_jupiter

# Tworzenie animacji zsynchronizowanej z rzeczywistym czasem
frames = int(T_moon / dt) + 1  # Liczba klatek na podstawie najdłuższego czasu (Księżyc)
interval = dt * 1000  # Interwał w milisekundach (0.02 s = 20 ms)
ani = FuncAnimation(fig, update, frames=frames, init_func=init, blit=False, interval=interval)

# Wyświetlenie animacji w Colabie
plt.close()
HTML(ani.to_html5_video())
```
