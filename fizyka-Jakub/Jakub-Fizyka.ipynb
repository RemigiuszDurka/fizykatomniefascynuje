# Równania Różniczkowe Ruchu Ciała Przymocowanego do Sprężyny

Notatka ta prezentuje matematyczny model ruchu oscylatora harmonicznego – ciała przymocowanego do sprężyny. Wyprowadzenie opiera się na podstawowych zasadach mechaniki klasycznej. W dalszej części dowiemy się, skąd wywodzi się dane równanie, czym są poszczególne jego elementy, a także jakie znaczenie mają warunki początkowe.

## 1. Ogólny Model Ruchu

### 1.1 Pochodzenie Równań

Model ruchu ciała przymocowanego do sprężyny opiera się na dwóch podstawowych zasadach:

- **Prawo Hooke'a**: Opisuje, że siła sprężystości wytwarzana przez sprężynę jest wprost proporcjonalna do wychylenia z położenia równowagi, ale skierowana przeciwnie do tego wychylenia. Matematycznie zapisujemy to równaniem:
  
  $$
  F = -kx,
  $$
  
  gdzie:
  - $ k $ – stała sprężystości, charakteryzująca sztywność sprężyny,
  - $ x $ – wychylenie ciała z położenia równowagi.
  
- **Druga Zasada Dynamiki Newtona**: Mówi, że przyspieszenie ciała jest proporcjonalne do wypadkowej siły działającej na nie oraz odwrotnie proporcjonalne do jego masy. Wyrażamy to równaniem:
  
  $$
  F = m \, a = m \, \frac{d^2 x}{dt^2},
  $$
  
  gdzie:
  - $ m $ – masa ciała,
  - $ a $ – przyspieszenie ciała, czyli druga pochodna przemieszczenia $ x(t) $ względem czasu $ t $.

Łącząc te dwa prawa, uzyskujemy równanie ruchu dla układu drgającego.

### 1.2 Równanie Ruchu

Podstawiając wyrażenie siły z prawa Hooke'a do równania Newtona, otrzymujemy:

$$
m \frac{d^2 x}{dt^2} + kx = 0.
$$

Jest to liniowe, jednorodne równanie różniczkowe drugiego rzędu, które opisuje ruch oscylacyjny.

### 1.3 Znaczenie Poszczególnych Elementów

- **Masa $m$**: Odpowiada za bezwładność ciała. Im większa masa, tym trudniej jest zmienić ruch ciała.
- **Stała sprężystości $k$**: Określa, jak mocno sprężyna "przywraca" ciało do położenia równowagi. Wyższa wartość oznacza większą siłę sprężystości przy danym wychyleniu.
- **Wychylenie $x(t)$**: Pozycja ciała w czasie, czyli jak daleko i w którą stronę od położenia równowagi znajduje się ciało.
- **Przyspieszenie $\frac{d^2 x}{dt^2}$**: Zmiana prędkości ciała, która jest bezpośrednio związana z działającą siłą.

### 1.4 Rozwiązanie Ogólne

Aby rozwiązać równanie różniczkowe, korzystamy z metody rozwiązań postaci wykładniczych lub trygonometrycznych. Rozwiązanie ogólne przedstawia się w postaci:

$$
x(t) = A \cos(\omega t) + B \sin(\omega t)
$$

lub alternatywnie z przesunięciem fazowym:

$$
x(t) = C \cos(\omega t + \phi),
$$

gdzie:
- $$ \omega = \sqrt{\frac{k}{m}} $$ – częstość kątowa drgań,
- $ A $, $ B $ (lub $
$ C $
$ i $ \phi $) – stałe wyznaczane na podstawie warunków początkowych.

### 1.5 Warunki Początkowe

Warunki początkowe są niezbędne do określenia jednoznacznego rozwiązania równania różniczkowego. Najczęściej przyjmujemy:
- $$ x(0) = x_0 $$ – wartość wychylenia w chwili $t = 0 $,
- $$ \dot{x}(0) = v_0 $$ – wartość prędkości w chwili $ t = 0 $.

Warunki te pozwalają nam wyznaczyć stałe $ A $ i $ B $:

1. Podstawiając $ t = 0 $ do ogólnego rozwiązania:

   $$
   x(0) = A \cos(0) + B \sin(0) = A \quad \Rightarrow \quad A = x_0.
   $$

2. Obliczając pochodną:

   $$
   \dot{x}(t) = -A \omega \sin(\omega t) + B \omega \cos(\omega t),
   $$
  
   a następnie przy $ t = 0 $:

   $$
   \dot{x}(0) = B \omega = v_0 \quad \Rightarrow \quad B = \frac{v_0}{\omega}.
   $$

Ostatecznie rozwiązanie z określonymi warunkami początkowymi ma postać:

$$
x(t) = x_0 \cos(\omega t) + \left(\frac{v_0}{\omega}\right) \sin(\omega t).
$$

---

## 2. Przypadek Szczególny:

### 2.1 Ustalone Parametry

Dla konkretnego przykładu przyjmujemy:
- Masa: $$m = 1\,\text{kg}$$
- Stała sprężystości: przykładowo $$k = 4 \, \text{N/m}$$ (można przyjąć dowolną wartość, ale ta ułatwia obliczenia)
- Wychylenie początkowe: $$x(0) = x_0 = 0.5\,\text{m}$$
- Prędkość początkowa: $$\dot{x}(0) = v_0 = 0 \,\text{m/s}$$

Wówczas:

$$
\omega = \sqrt{\frac{k}{m}} = \sqrt{\frac{4}{1}} = 2\,\text{s}^{-1}.
$$

### 2.2 Obliczenia dla Wybranych Warunków

Podstawiamy warunki do rozwiązania ogólnego:

1. Stała $$A$$
   
   $$
   A = x_0 = 0.5\,\text{m}.
   $$

2. Stała $$B$$

   $$
   B = \frac{v_0}{\omega} = \frac{0}{2} = 0.
   $$

Rozwiązanie przyjmuje postać:

$$
x(t) = 0.5 \cos(2t).
$$

### 2.3 Interpretacja Wyniku

Otrzymane rozwiązanie:

$$
x(t) = 0.5 \cos(2t)
$$

oznacza, że ciało wykonuje ruch oscylacyjny o:
- **Amplitudzie**: $$0.5\,\text{m}$$ – maksymalne wychylenie z położenia równowagi,
- **Częstości kątowej**: $$\omega = 2\,\text{s}^{-1}$$.

Okres oscylacji $T$, czyli czas jednego pełnego drgania, wynosi:

$$
T = \frac{2\pi}{\omega} = \frac{2\pi}{2} = \pi\,\text{s}.
$$

Oznacza to, że ciało wraca do swojego początkowego stanu co $$\pi$$ sekund, a energia układu cyklicznie zamienia się między energią kinetyczną a potencjalną.

---

## 3. Podsumowanie i Interpretacja

### 3.1 Skąd się biorą poszczególne składniki?

- **Prawo Hooke'a** wyprowadza się eksperymentalnie i opisuje zachowanie sprężyny (oraz innych sprężystych elementów). Oparte jest na obserwacji liniowej zależności między siłą a wychyleniem.
- **Druga zasada dynamiki Newtona** stanowi fundament mechaniki klasycznej i opisuje dynamikę ciał.
- Łącząc oba te prawa, otrzymujemy równanie ruchu, które jest powszechnie stosowane do opisu systemów oscylacyjnych, od prostych układów mechanicznych po bardziej złożone układy fizyczne.

### 3.2 Czym są warunki początkowe?

Warunki początkowe, czyli $x(0) = x_0$ i $\dot{x}(0) = v_0$, to wartości określające stan układu w momencie rozpoczęcia obserwacji (w chwili $t = 0$). Dzięki nim:
- Możemy jednoznacznie określić rozwiązanie równania różniczkowego – bez nich otrzymalibyśmy ogólny zbiór rozwiązań.
- Określają, jak układ rozpocznie swoje zachowanie – czy ciało zostanie nadane z początkowym wychyleniem, czy też z dodatkową prędkością.
- Umożliwiają modelowanie rzeczywistych sytuacji, gdzie stan początkowy jest znany (np. wypuszczenie ciała z określonej pozycji przy zerowej prędkości lub z początkowym pędem).

### 3.3 Znaczenie Modelu

Model oscylatora harmonicznego jest niezwykle uniwersalny i znajduje zastosowanie w wielu dziedzinach:
- W **fizyce** do opisu drgań mechanicznych, elektrycznych drgań w obwodach LC czy w modelach kwantowych.
- W **inżynierii** przy projektowaniu systemów drgań i tłumienia (np. w zawieszeniach pojazdów).
- W **matematyce** jako klasyczny przykład liniowego równania różniczkowego o stałych współczynnikach.

Takie podejście nie tylko pozwala na teoretyczne zrozumienie dynamiki układu, ale również na praktyczne rozwiązania w symulacjach i eksperymentach.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry układu
m = 1.0    # masa [kg]
k = 4.0    # stała sprężystości [N/m]
x0 = 0.5   # położenie początkowe [m]
v0 = 0.0   # prędkość początkowa [m/s]

# Obliczenia pomocnicze
omega = np.sqrt(k/m)       # częstość drgań własnych
T = 2*np.pi/omega          # okres drgań
t_max = 3*T                # czas symulacji = 3 okresy

# Generowanie danych
t = np.linspace(0, t_max, 1000)
x = x0 * np.cos(omega * t)  # rozwiązanie analityczne

# Konfiguracja wykresu
plt.figure(figsize=(12, 6))
plt.plot(t, x, 'b-', linewidth=2, label=f'x(t) = {x0} cos({omega:.1f}t)')

# Zaznaczenie amplitudy
plt.axhline(y=x0, color='r', linestyle='--', alpha=0.5, label='Amplituda')
plt.axhline(y=-x0, color='r', linestyle='--', alpha=0.5)

# Zaznaczenie okresów
for i in range(1, 4):
    plt.axvline(x=i*T, color='g', linestyle=':', alpha=0.5, label='Okres' if i == 1 else "")

# Dodanie opisów
plt.title('Ruch harmoniczny ciała na sprężynie', fontsize=14)
plt.xlabel('Czas [s]', fontsize=12)
plt.ylabel('Położenie [m]', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(np.arange(0, t_max+0.1, T/2))
plt.xlim([0, t_max])

# Wyświetlenie wykresu
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle
from matplotlib import rcParams
from IPython.display import HTML, display

# Ustawienia stylu
rcParams['font.family'] = 'serif'
plt.style.use('ggplot')

# Parametry układu
params = {
    'm': 1.0,      # masa [kg]
    'k': 4.0,      # stała sprężystości [N/m]
    'x0': 0.5,     # położenie początkowe [m]
    'v0': 0.0      # prędkość początkowa [m/s]
}

# Obliczenia fizyczne
omega = np.sqrt(params['k'] / params['m'])  # częstość drgań
T = 2 * np.pi / omega                       # okres drgań
t_max = 3 * T                               # czas symulacji
energy = 0.5 * params['k'] * params['x0'] ** 2  # energia całkowita

# Przygotowanie danych czasowych
t = np.linspace(0, t_max, 300)
x = params['x0'] * np.cos(omega * t)         # rozwiązanie analityczne
v = -params['x0'] * omega * np.sin(omega * t)  # prędkość

# Konfiguracja figury za pomocą GridSpec – 3 panele: główny, energia oraz portret fazowy
fig = plt.figure(figsize=(14, 7))
gs = fig.add_gridspec(2, 2, width_ratios=[3, 1])
ax_main = fig.add_subplot(gs[:, 0])
ax_energy = fig.add_subplot(gs[0, 1])
ax_phase = fig.add_subplot(gs[1, 1])

# Główny panel - układ fizyczny
ax_main.set_xlim(-2, 2)
ax_main.set_ylim(-0.7, 1.2)
ax_main.set_aspect('equal')
ax_main.set_title('Animacja ruchu harmonicznego', fontsize=14, pad=20)
ax_main.set_xlabel('Położenie [m]', fontsize=12)
ax_main.grid(True, alpha=0.4)

# Elementy wizualne – nie ma już dodatkowego obrazu (np. ściany)
n_coils = 10
spring, = ax_main.plot([], [], 'b-', lw=2, alpha=0.8)
mass = Circle((params['x0'], 0), 0.15, fc='#e74c3c', ec='black', lw=1.5)
equilibrium = ax_main.axvline(0, color='gray', ls=':', alpha=0.7)
ax_main.add_patch(mass)

# Teksty informacyjne
time_text = ax_main.text(0.02, 0.95, '', transform=ax_main.transAxes,
                         bbox=dict(facecolor='white', alpha=0.8))
param_text = ax_main.text(1.02, 0.95,
    f"m = {params['m']} kg\nk = {params['k']} N/m\nω = {omega:.2f} rad/s\nT = {T:.2f} s",
    transform=ax_main.transAxes, ha='left', va='top',
    bbox=dict(facecolor='white', alpha=0.7))

# Wykres energii
ax_energy.set_title('Energia układu', fontsize=12)
ax_energy.set_xlim(0, t_max)
ax_energy.set_ylim(0, energy * 1.2)
ax_energy.set_ylabel('Energia [J]')
energy_line, = ax_energy.plot([], [], 'g-', lw=2, label='Energia całkowita')
pe_line, = ax_energy.plot([], [], 'b--', lw=1.5, alpha=0.7, label='Energia potencjalna')
ke_line, = ax_energy.plot([], [], 'r--', lw=1.5, alpha=0.7, label='Energia kinetyczna')
ax_energy.legend(fontsize=9)

# Portret fazowy
ax_phase.set_title('Przestrzeń fazowa', fontsize=12)
ax_phase.set_xlabel('Położenie [m]')
ax_phase.set_ylabel('Prędkość [m/s]')
phase_line, = ax_phase.plot([], [], 'm-', lw=1.5, alpha=0.7)
phase_point, = ax_phase.plot([], [], 'mo', markersize=8)
ax_phase.grid(True, alpha=0.3)

# Funkcja inicjalizująca animację
def init():
    spring.set_data([], [])
    time_text.set_text('')
    mass.center = (params['x0'], 0)
    return spring, mass, time_text

# Funkcja animująca – aktualizuje położenie sprężyny, masy, wykresy energii oraz portret fazowy
def animate(i):
    # Sprężyna: rysujemy ją między punktem stałym (x=-1.7) a aktualnym położeniem masy
    spring_x = np.linspace(-1.7, x[i], 100)
    spring_y = 0.1 * np.sin(np.linspace(0, n_coils * 2 * np.pi, 100))
    spring.set_data(spring_x, spring_y)

    # Aktualizacja położenia masy
    mass.center = (x[i], 0)

    # Aktualizacja tekstu informacyjnego
    time_text.set_text(f'Czas: {t[i]:.2f} s\nx = {x[i]:.2f} m\nv = {v[i]:.2f} m/s')

    # Obliczenia energii
    pe = 0.5 * params['k'] * x[i] ** 2  # energia potencjalna
    ke = 0.5 * params['m'] * v[i] ** 2    # energia kinetyczna
    energy_total = pe + ke

    energy_line.set_data(t[:i], energy_total * np.ones(i))
    pe_line.set_data(t[:i], [0.5 * params['k'] * x[j] ** 2 for j in range(i)])
    ke_line.set_data(t[:i], [0.5 * params['m'] * v[j] ** 2 for j in range(i)])

    # Aktualizacja portretu fazowego
    phase_line.set_data(x[:i], v[:i])
    phase_point.set_data([x[i]], [v[i]])
    ax_phase.set_xlim(min(x)-0.1, max(x)+0.1)
    ax_phase.set_ylim(min(v)-0.1, max(v)+0.1)

    return (spring, mass, time_text, energy_line, pe_line,
            ke_line, phase_line, phase_point)

# Tworzenie animacji
ani = FuncAnimation(fig, animate, frames=len(t), init_func=init,
                    interval=20, blit=True, repeat_delay=1000)

# Legenda w głównym panelu
ax_main.legend([spring, mass, equilibrium],
               ['Sprężyna', f'Ciało (m={params["m"]}kg)', 'Położenie równowagi'],
               loc='lower right', fontsize=10)

plt.tight_layout()

# Aby uniknąć wyświetlenia statycznego obrazu (ostatniej klatki) pod animacją,
# wywołujemy display() dla animacji, a następnie zamykamy figurę.
html_anim = HTML(ani.to_html5_video())
display(html_anim)
plt.close(fig)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle
from matplotlib import rcParams
from IPython.display import HTML, display

# Ustawienia stylu
rcParams['font.family'] = 'serif'
plt.style.use('ggplot')

# Parametry układu
params = {
    'm': 1.0,      # masa [kg]
    'k': 4.0,      # stała sprężystości [N/m]
    'x0': 0.5,     # położenie początkowe [m]
    'v0': 0.0,     # prędkość początkowa [m/s]
    'b': 0.2       # współczynnik tłumienia [kg/s]
}

# Obliczenia teoretyczne:
omega = np.sqrt(params['k'] / params['m'])  # charakterystyczna częstość (dla oscylatora bez tłumienia)
T = 2 * np.pi / omega                       # okres drgań (w warunkach nie tłumionych)
t_max = 9 * T                               # czas symulacji
n_frames = 300                              # liczba klatek animacji
dt = t_max / n_frames                       # krok czasowy
t = np.linspace(0, t_max, n_frames)

# Rozwiązanie numeryczne dla oscylatora tłumionego (metoda Euler-Cromer):
x = np.zeros_like(t)
v = np.zeros_like(t)
x[0] = params['x0']
v[0] = params['v0']

for i in range(1, len(t)):
    # Równanie ruchu: m * a = - k * x - b * v
    a = -(params['k'] * x[i-1] + params['b'] * v[i-1]) / params['m']
    v[i] = v[i-1] + a * dt
    x[i] = x[i-1] + v[i] * dt

# Obliczenie energii:
pe = 0.5 * params['k'] * x**2         # energia potencjalna
ke = 0.5 * params['m'] * v**2           # energia kinetyczna
energy_total = pe + ke                # energia całkowita
energy_max = np.max(pe)                # maksymalna energia potencjalna dla skali wykresu

# Konfiguracja figury: 3 panele – główny (ruch na sprężynie), wykres energii oraz portret fazowy
fig = plt.figure(figsize=(14, 7))
gs = fig.add_gridspec(2, 2, width_ratios=[3, 1])
ax_main = fig.add_subplot(gs[:, 0])
ax_energy = fig.add_subplot(gs[0, 1])
ax_phase = fig.add_subplot(gs[1, 1])

# Panel główny – animacja ruchu na sprężynie
ax_main.set_xlim(-2, 2)
ax_main.set_ylim(-0.7, 1.2)
ax_main.set_aspect('equal')
ax_main.set_title('Ruch ciała na sprężynie z siłą oporu', fontsize=14, pad=20)
ax_main.set_xlabel('Położenie [m]', fontsize=12)
ax_main.grid(True, alpha=0.4)

# Elementy wizualne w panelu głównym:
n_coils = 10  # liczba "zwojów" sprężyny
spring, = ax_main.plot([], [], 'b-', lw=2, alpha=0.8)
mass = Circle((params['x0'], 0), 0.15, fc='#e74c3c', ec='black', lw=1.5)
equilibrium = ax_main.axvline(0, color='gray', ls=':', alpha=0.7)
ax_main.add_patch(mass)
time_text = ax_main.text(0.02, 0.95, '', transform=ax_main.transAxes,
                         bbox=dict(facecolor='white', alpha=0.8))
param_text = ax_main.text(1.02, 0.95,
    f"m = {params['m']} kg\nk = {params['k']} N/m\nb = {params['b']} kg/s\nω ≈ {omega:.2f} rad/s\nT ≈ {T:.2f} s",
    transform=ax_main.transAxes, ha='left', va='top',
    bbox=dict(facecolor='white', alpha=0.7))

# Konfiguracja wykresu energii
ax_energy.set_title('Energia układu', fontsize=12)
ax_energy.set_xlim(0, t_max)
ax_energy.set_ylim(0, energy_max * 1.2)
ax_energy.set_ylabel('Energia [J]')
energy_line, = ax_energy.plot([], [], 'g-', lw=2, label='Energia całkowita')
pe_line, = ax_energy.plot([], [], 'b--', lw=1.5, alpha=0.7, label='Energia potencjalna')
ke_line, = ax_energy.plot([], [], 'r--', lw=1.5, alpha=0.7, label='Energia kinetyczna')
ax_energy.legend(fontsize=9)

# Konfiguracja portretu fazowego
ax_phase.set_title('Przestrzeń fazowa', fontsize=12)
ax_phase.set_xlabel('Położenie [m]')
ax_phase.set_ylabel('Prędkość [m/s]')
phase_line, = ax_phase.plot([], [], 'm-', lw=1.5, alpha=0.7)
phase_point, = ax_phase.plot([], [], 'mo', markersize=8)
ax_phase.grid(True, alpha=0.3)

# Funkcja inicjalizująca animację
def init():
    spring.set_data([], [])
    time_text.set_text('')
    mass.center = (params['x0'], 0)
    energy_line.set_data([], [])
    pe_line.set_data([], [])
    ke_line.set_data([], [])
    phase_line.set_data([], [])
    phase_point.set_data([], [])
    return (spring, mass, time_text, energy_line, pe_line, ke_line, phase_line, phase_point)

# Funkcja animująca – aktualizuje ruch ciała, energię oraz portret fazowy
def animate(i):
    # Rysowanie sprężyny: od stałego punktu (x = -1.7) do aktualnego położenia ciała
    spring_x = np.linspace(-1.7, x[i], 100)
    spring_y = 0.1 * np.sin(np.linspace(0, n_coils * 2 * np.pi, 100))
    spring.set_data(spring_x, spring_y)

    # Aktualizacja położenia ciała
    mass.center = (x[i], 0)

    # Aktualizacja tekstu informacyjnego: czas, położenie, prędkość
    time_text.set_text(f'Czas: {t[i]:.2f} s\nx = {x[i]:.2f} m\nv = {v[i]:.2f} m/s')

    # Uaktualnienie wykresów energii
    energy_line.set_data(t[:i], energy_total[:i])
    pe_line.set_data(t[:i], pe[:i])
    ke_line.set_data(t[:i], ke[:i])

    # Aktualizacja portretu fazowego tylko, gdy mamy już dane (i>0)
    if i > 0:
        phase_line.set_data(x[:i], v[:i])
        phase_point.set_data([x[i]], [v[i]])
        ax_phase.set_xlim(np.min(x[:i]) - 0.1, np.max(x[:i]) + 0.1)
        ax_phase.set_ylim(np.min(v[:i]) - 0.1, np.max(v[:i]) + 0.1)

    return (spring, mass, time_text, energy_line, pe_line, ke_line, phase_line, phase_point)

# Tworzenie animacji
ani = FuncAnimation(fig, animate, frames=n_frames, init_func=init,
                    interval=20, blit=True, repeat_delay=1000)

plt.tight_layout()

# Wyświetlenie animacji (HTML5 w środowisku Notebook)
html_anim = HTML(ani.to_html5_video())
display(html_anim)
plt.close(fig)
```

# Ciało zawieszone na sprężynie – model teoretyczny

W tej notatce rozważamy układ, w którym ciało o masie $m$ zawieszone jest na sprężynie o stałej sprężystości $k$, umieszczonej pionowo (wzdłuż osi $y$). Analizujemy ruch ciała, uwzględniając zarówno siłę sprężystości, jak i wpływ siły grawitacji (przyspieszenie $g$).

## 1. Opis układu i sił działających

Zakładamy, że:
- Oś $y$ jest skierowana w dół.
- $y = 0$ oznacza położenie, przy którym sprężyna jest niespnięta (długość swobodna $L_0$).
- Dla dowolnego położenia $y$ działają następujące siły:
  - Siła sprężystości: $F_{\text{spr}} = -k \, (y - L_0)$,
  - Siła ciężkości (grawitacji): $F_g = m \, g$ (skierowana w dół).

## 2. Wyznaczenie punktu równowagi

W punkcie równowagi, przy którym ciało pozostaje w spoczynku ($\ddot{y} = 0$), suma sił musi być równa zero. Zatem mamy:
$$
-k \, (y_{\text{eq}} - L_0) + m \, g = 0.
$$
Rozwiązując względem $y_{\text{eq}}$, otrzymujemy:
$$
y_{\text{eq}} = L_0 + \frac{m \, g}{k}.
$$
Grawitacja powoduje, że równowagowe położenie ciała jest przesunięte o $\frac{m \, g}{k}$ w dół względem pozycji $L_0$.

## 3. Równanie ruchu – podejście klasyczne

Korzystając z II zasady dynamiki Newtona, mamy:
$$
m \, \ddot{y} = \sum F.
$$
Podstawiając siły działające na ciało, otrzymujemy:
$$
m \, \ddot{y} = -k \, (y - L_0) + m \, g.
$$
Możemy zapisać równanie w postaci:
$$
m \, \ddot{y} + k \, y = k \, L_0 + m \, g.
$$

## 4. Redukcja równania przez zmianę zmiennej

Wprowadzamy zmienną opisującą wychylenie względem położenia równowagi:
$$
x = y - y_{\text{eq}}.
$$
Skoro $y = x + y_{\text{eq}}$, po podstawieniu do równania ruchu uzyskujemy:
$$
m \, \ddot{x} + k \, (x + y_{\text{eq}}) = k \, L_0 + m \, g.
$$
Znając zależność:
$$
y_{\text{eq}} = L_0 + \frac{m \, g}{k},
$$
podstawiamy i zauważamy, że wyrazy niezależne od $x$ znoszą się, co prowadzi do jednorodnego równania:
$$
m \, \ddot{x} + k \, x = 0.
$$
Możemy to zapisać również jako:
$$
\ddot{x} + \omega^2 \, x = 0, \quad \text{gdzie} \quad \omega^2 = \frac{k}{m}.
$$

## 5. Rozwiązanie równania ruchu

Równanie:
$$
\ddot{x} + \omega^2 \, x = 0
$$
opisuje ruch harmoniczny, którego ogólne rozwiązanie można zapisać jako:
$$
x(t) = A \cos(\omega t) + B \sin(\omega t),
$$
gdzie $A$ i $B$ są stałymi wyznaczonymi z warunków początkowych.

W związku z tym, całkowite położenie ciała $y(t)$ możemy opisać równaniem:
$$
y(t) = y_{\text{eq}} + A \cos(\omega t) + B \sin(\omega t),
$$
gdzie punkt równowagi wynosi:
$$
y_{\text{eq}} = L_0 + \frac{m \, g}{k},
$$
a częstotliwość drgań określa:
$$
\omega = \sqrt{\frac{k}{m}}.
$$

## 6. Wpływ grawitacji na ruch

Grawitacja wprowadza dwa istotne efekty:
- **Przesunięcie punktu równowagi:** Grawitacja powoduje, że stan równowagi nie znajduje się przy $y = L_0$, lecz przy:
  $$
  y_{\text{eq}} = L_0 + \frac{m \, g}{k}.
  $$
- **Niezmienność częstotliwości oscylacji:** Po przejściu do układu odniesienia względem położenia równowagi, równanie ruchu przybiera formę jednorodną,
  $$
  \ddot{x} + \omega^2 \, x = 0,
  $$
  co oznacza, że częstotliwość drgań:
  $$
  \omega = \sqrt{\frac{k}{m}},
  $$
  nie zależy od grawitacji. Grawitacja jedynie przesuwa poziom odniesienia, lecz nie wpływa na dynamikę oscylacji.

---

## Podsumowanie

Rozważając ruch ciała zawieszonego na sprężynie w polu grawitacyjnym, otrzymujemy równanie ruchu:
$$
m \, \ddot{y} = -k \, (y - L_0) + m \, g,
$$
które przy przesunięciu zmiennej do położenia równowagi prowadzi do klasycznego równania harmonicznego:
$$
\ddot{x} + \frac{k}{m} \, x = 0.
$$
Grawitacja wpływa wyłącznie na przesunięcie położenia równowagi (do $y_{\text{eq}} = L_0 + \frac{m \, g}{k}$), natomiast częstotliwość oscylacji pozostaje określona przez zależność:
$$
\omega = \sqrt{\frac{k}{m}}.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Circle
from matplotlib import rcParams
from IPython.display import HTML, display

# Ustawienia stylu
rcParams['font.family'] = 'serif'
plt.style.use('ggplot')

# Parametry układu
params = {
    'm': 1.0,      # masa [kg]
    'k': 4.0,      # stała sprężystości [N/m]
    'b': 0.5,      # współczynnik tłumienia [kg/s]
    'g': 9.81,     # przyspieszenie grawitacyjne [m/s²]
    'x0': 0.5,     # początkowe wychylenie (odchylenie względem równowagi) [m]
    'v0': 0.0,     # prędkość początkowa [m/s]
    'L0': 1.0      # naturalna długość sprężyny (bez obciążenia) [m]
}

# Położenia równowagi:
L0 = params['L0']
# Naturalne położenie równowagi (sprężyna nieobciążona)
y_eq_no_g = L0
# Położenie równowagi przy grawitacji: L₀ + mg/k
y_eq = L0 + params['m'] * params['g'] / params['k']

# Przyjmujemy analizę drgań wokół położenia równowagi przy grawitacji.
# Definiujemy zmienną: x = y - y_eq.
# Warunki początkowe: x(0)=x0, v(0)=v0.
omega_eff = np.sqrt(params['k'] / params['m'])
T = 2 * np.pi / omega_eff
t_max = 3 * T
n_frames = 300
dt = t_max / n_frames
t = np.linspace(0, t_max, n_frames)

# Symulacja metodą Euler-Cromer dla wychylenia x (odchylenia od równowagi przy grawitacji)
x = np.zeros_like(t)
v = np.zeros_like(t)
x[0] = params['x0']
v[0] = params['v0']

for i in range(1, len(t)):
    a = -(params['k'] * x[i-1] + params['b'] * v[i-1]) / params['m']
    v[i] = v[i-1] + a * dt
    x[i] = x[i-1] + v[i] * dt

# Rzeczywista pozycja ciała (układ pionowy): y = x + y_eq.
y = x + y_eq

# Obliczenie energii oscylatora (drgania względem równowagi)
ke = 0.5 * params['m'] * v**2          # energia kinetyczna
pe = 0.5 * params['k'] * x**2           # energia potencjalna sprężyny
energy_total = ke + pe
energy_max = np.max(energy_total)

# Konfiguracja figury: trzy panele – ruch ciała, energia oraz portret fazowy
fig = plt.figure(figsize=(14, 7))
gs = fig.add_gridspec(2, 2, width_ratios=[3, 1])
ax_main = fig.add_subplot(gs[:, 0])
ax_energy = fig.add_subplot(gs[0, 1])
ax_phase = fig.add_subplot(gs[1, 1])

# PANEL GŁÓWNY – Animacja ruchu ciała na sprężynie (układ pionowy)
ax_main.set_xlim(-1, 1)
ax_main.set_ylim(-0.5, y_eq + params['x0'] + 1)
ax_main.invert_yaxis()  # odwracamy oś y, aby punkt mocowania (0,0) był na górze
ax_main.set_title('Ruch ciała zawieszonego na sprężynie\nz grawitacją i siłą oporu', fontsize=14, pad=20)
ax_main.set_xlabel('Pozycja w poziomie [m]', fontsize=12)
ax_main.set_ylabel('Pozycja w pionie [m]', fontsize=12)
ax_main.grid(True, alpha=0.4)

# Elementy głównego panelu:
n_coils = 10
spring_line, = ax_main.plot([], [], 'b-', lw=2, alpha=0.8)
mass = Circle((0, y[0]), 0.15, fc='#e74c3c', ec='black', lw=1.5)
ax_main.add_patch(mass)
# Linia pokazująca położenie równowagi przy grawitacji
eq_line = ax_main.axhline(y_eq, color='gray', ls='--', lw=1.5, label=f"yₑq (z grawitacją): {y_eq:.2f} m")
# Linia pokazująca naturalne (nieobciążone) położenie równowagi
eq_line_no_g = ax_main.axhline(y_eq_no_g, color='purple', ls='--', lw=1.5, label=f"y₀ (bez grawitacji): {y_eq_no_g:.2f} m")

time_text = ax_main.text(0.02, 0.05, '', transform=ax_main.transAxes,
                         bbox=dict(facecolor='white', alpha=0.8))
param_text = ax_main.text(1.02, 0.05,
    (f"m = {params['m']} kg\nk = {params['k']} N/m\nb = {params['b']} kg/s\n"
     f"g = {params['g']} m/s²\nx₀ = {params['x0']} m\nv₀ = {params['v0']} m/s\nL₀ = {params['L0']} m"),
    transform=ax_main.transAxes, ha='left', va='bottom',
    bbox=dict(facecolor='white', alpha=0.7))
ax_main.legend(loc='upper right', fontsize=10)

# PANEL ENERGII
ax_energy.set_title('Energia układu', fontsize=12)
ax_energy.set_xlim(0, t_max)
ax_energy.set_ylim(0, energy_max * 1.2)
ax_energy.set_ylabel('Energia [J]')
energy_line, = ax_energy.plot([], [], 'g-', lw=2, label='Energia całkowita')
ke_line, = ax_energy.plot([], [], 'r--', lw=1.5, alpha=0.7, label='Energia kinetyczna')
pe_line, = ax_energy.plot([], [], 'b--', lw=1.5, alpha=0.7, label='Energia sprężyny')
ax_energy.legend(fontsize=9)

# PANEL PRZESTRZENI FAZOWEJ
ax_phase.set_title('Przestrzeń fazowa', fontsize=12)
ax_phase.set_xlabel('Wychylenie od równowagi x [m]')
ax_phase.set_ylabel('Prędkość [m/s]')
phase_line, = ax_phase.plot([], [], 'm-', lw=1.5, alpha=0.7)
phase_point, = ax_phase.plot([], [], 'mo', markersize=8)
ax_phase.grid(True, alpha=0.3)

# Funkcja inicjalizująca animację
def init():
    spring_line.set_data([], [])
    time_text.set_text('')
    mass.center = (0, y[0])
    energy_line.set_data([], [])
    ke_line.set_data([], [])
    pe_line.set_data([], [])
    phase_line.set_data([], [])
    phase_point.set_data([], [])
    return spring_line, mass, time_text, energy_line, ke_line, pe_line, phase_line, phase_point

# Funkcja animująca
def animate(i):
    # Rysowanie sprężyny: punkty pomiędzy punktem mocowania (0,0) a masą (0, y[i])
    coil_y = np.linspace(0, y[i], 100)
    amplitude = 0.1
    coil_x = amplitude * np.sin(np.linspace(0, n_coils * 2 * np.pi, 100))
    spring_line.set_data(coil_x, coil_y)

    # Aktualizacja pozycji ciała
    mass.center = (0, y[i])

    # Aktualizacja tekstu informacyjnego
    time_text.set_text(f'Czas: {t[i]:.2f} s\nx = {x[i]:.2f} m\nv = {v[i]:.2f} m/s')

    # Aktualizacja wykresów energii
    energy_line.set_data(t[:i], energy_total[:i])
    ke_line.set_data(t[:i], ke[:i])
    pe_line.set_data(t[:i], pe[:i])

    # Aktualizacja portretu fazowego
    phase_line.set_data(x[:i], v[:i])
    phase_point.set_data([x[i]], [v[i]])
    if i > 0:
        ax_phase.set_xlim(np.min(x[:i]) - 0.1, np.max(x[:i]) + 0.1)
        ax_phase.set_ylim(np.min(v[:i]) - 0.1, np.max(v[:i]) + 0.1)

    return spring_line, mass, time_text, energy_line, ke_line, pe_line, phase_line, phase_point

# Tworzenie animacji – ustawiamy blit=False dla poprawnej aktualizacji wszystkich paneli
ani = FuncAnimation(fig, animate, frames=n_frames, init_func=init,
                    interval=20, blit=False, repeat_delay=1000)

plt.tight_layout()

# Wyświetlenie animacji w formacie HTML5 (np. w Jupyter Notebook)
html_anim = HTML(ani.to_html5_video())
display(html_anim)
plt.close(fig)
```

# Krzywa Lissajous

Krzywe Lissajous to figury powstające w wyniku złożenia dwóch drgań harmonicznych w dwóch prostopadłych kierunkach. Są wykorzystywane m.in. w analizie sygnałów, fizyce, elektronice, akustyce i grafice komputerowej.

## Wzory parametryczne

Krzywą Lissajous opisuje się za pomocą równań parametrycznych:

$$
x(t) = A \sin(at + \delta)
$$

$$
y(t) = B \sin(bt)
$$

Gdzie:
- $x(t)$ – położenie w osi poziomej w danym momencie czasu $t$,
- $y(t)$ – położenie w osi pionowej w danym momencie czasu $t$,
- $A$, $B$ – amplitudy drgań w kierunkach $x$ i $y$ (czyli maksymalne wychylenie od środka),
- $a$, $b$ – częstotliwości drgań w kierunkach $x$ i $y$,
- $\delta$ – różnica fazowa między drganiami (czyli przesunięcie jednego z sygnałów względem drugiego),
- $t$ – czas, pełniący rolę parametru.

## Na czym to polega?

Krzywa Lissajous powstaje, gdy punkt porusza się jednocześnie w dwóch prostopadłych kierunkach – w poziomie (oś $x$) i w pionie (oś $y$) – zgodnie z ruchem harmonicznym (sinusoidalnym). Ruch ten może mieć różne częstotliwości i amplitudy w każdej z osi oraz pewne przesunięcie fazowe.

## Wpływ stosunku częstotliwości $\frac{a}{b}$

Kształt krzywej Lissajous jest silnie zależny od stosunku częstotliwości $a$ i $b$. Rozważmy różne przypadki:

### 1. $\frac{a}{b} = 1$ (częstotliwości równe)
- Przy $\delta = 0$ lub $\pi$ krzywa jest linią prostą (ukośną).
- Przy $\delta = \frac{\pi}{2}$ – krzywa jest okręgiem (gdy $A = B$) lub elipsą (gdy $A \ne B$).

### 2. $\frac{a}{b} = \frac{m}{n}$ (stosunek wymierny)
- Krzywa jest zamknięta i powtarzalna.
- Kształt składa się z $n$ „pętli” w poziomie i $m$ w pionie.
- Przykład: dla $a = 3$, $b = 2$, otrzymujemy krzywą z 2 pętlami w poziomie i 3 w pionie.
- Liczba przecięć i symetria figury zależą od wartości $\delta$.

### 3. $\frac{a}{b}$ jest liczbą niewymierną (np. $\sqrt{2}$)
- Krzywa nigdy się nie zamyka.
- Poruszający się punkt nigdy nie wraca dokładnie do tego samego miejsca.
- Krzywa z czasem „wypełnia” obszar, tworząc strukturę przypominającą szum lub gęsto zapełnioną powierzchnię.

## Wpływ przesunięcia fazowego $\delta$
- $\delta$ kontroluje, jak bardzo fale są przesunięte względem siebie.
- Dla $\delta = 0$ obie fale zaczynają się równocześnie – krzywa ma najwyższą symetrię.
- Dla $\delta = \frac{\pi}{2}$ często powstają elipsy, okręgi lub bardziej złożone figury.
- Zmiana $\delta$ wpływa na „skręcenie” figury i jej orientację.

## Wpływ amplitud $A$ i $B$

Amplitudy $A$ i $B$ określają maksymalne wychylenie ruchu w osi poziomej ($x$) i pionowej ($y$). Innymi słowy, są to „rozmiary” drgań w każdej z osi. Wpływają one głównie na **kształt geometryczny** figury, a nie na to, czy krzywa się zamyka.

### 1. $A = B$
- Krzywa jest **symetryczna względem przekątnej**, gdy $\delta = \frac{\pi}{2}$.
- Figury mają **jednakowy zasięg** w poziomie i pionie – są „proporcjonalne”.
- Przykład: może powstać **okrąg** lub idealna symetryczna ósemka.

### 2. $A \ne B$
- Krzywa jest **rozciągnięta** w jednym z kierunków.
- Jeśli $A > B$, figura będzie szersza niż wyższa.
- Jeśli $A < B$, figura będzie wyższa niż szersza.
- Zachowana jest ogólna struktura krzywej, ale zmieniają się jej **wymiary i proporcje**.

### 3. $A = 0$ lub $B = 0$
- Jeśli $A = 0$, to $x(t) = 0$ – punkt porusza się tylko w pionie, czyli wzdłuż linii pionowej.
- Jeśli $B = 0$, to $y(t) = 0$ – punkt porusza się tylko w poziomie, czyli wzdłuż linii poziomej.
- W obu przypadkach nie powstaje żadna figura – tylko prosta linia wzdłuż jednej osi.

### 4. $A = B = 0$
- Punkt jest nieruchomy – nie ma żadnego ruchu ani figury.

### Podsumowanie:
Amplitudy odpowiadają za **rozmiar** i **proporcje** figury:
- Zmieniamy amplitudy → zmieniamy kształt (np. elipsa bardziej spłaszczona).
- Zmieniamy tylko częstotliwości → zmieniamy liczbę pętli, symetrię i złożoność wzoru.

## Wpływ przesunięcia fazowego $\delta$

Przesunięcie fazowe $\delta$ określa, o ile jedna fala (np. w osi $x$) jest opóźniona względem drugiej (np. w osi $y$). To przesunięcie wpływa na **orientację**, **symetrię**, a w niektórych przypadkach także na **kształt** figury Lissajous.

Rozważmy różne przypadki:

### 1. $\delta = 0$
- Obie funkcje sinusoidalne startują równocześnie.
- Dla $a = b$ i $A = B$, krzywa jest prostą rosnącą: $y = x$.
- Dla $a \ne b$ lub $A \ne B$, krzywa może być złożoną figurą, ale nadal bardzo symetryczną.
- Figura jest **symetryczna względem osi poziomej i pionowej**.

### 2. $\delta = \pi$
- Obie fale są w przeciwfazie – gdy jedna osiąga maksimum, druga minimum.
- Dla $a = b$ i $A = B$, krzywa to prosta malejąca: $y = -x$.
- Zmienia się kierunek „nachylenia” krzywej.

### 3. $\delta = \frac{\pi}{2}$
- Fale są przesunięte o ćwierć okresu – maksimum jednej przypada na zero drugiej.
- Dla $a = b$ i $A = B$, powstaje **okrąg**.
- Dla $A \ne B$, powstaje **elipsa**.
- Figura jest wtedy **maksymalnie „zakręcona”** – nie przecina osi, lecz je otacza.

### 4. $\delta = \frac{3\pi}{2}$
- Podobnie jak przy $\frac{\pi}{2}$, ale z odwrotną orientacją elipsy.
- Kierunek „skręcenia” figury jest przeciwny – np. przeciwny kierunek obiegu.

### 5. $\delta$ dowolne (ogólny przypadek)
- Zmienność $\delta$ od $0$ do $2\pi$ powoduje:
  - **obrót figury** w przestrzeni,
  - **zmianę punktów przecięcia z osiami**,
  - **zmianę liczby symetrii** (często je łamie),
  - przechodzenie figury przez stany: elipsa → ósemka → „węzeł” → rozciągnięta spirala.

### 6. $\delta$ a zamknięcie figury
- Jeśli $\frac{a}{b}$ jest wymierne, zmiana $\delta$ **nie wpływa na to, czy figura się zamknie**, ale wpływa na jej kształt.
- Przy niewymiernym $\frac{a}{b}$ figura nigdy się nie zamknie, niezależnie od wartości $\delta$.

### Podsumowanie:
- $\delta = 0$: prosta wzdłuż przekątnej (przy $a = b$),
- $\delta = \frac{\pi}{2}$: okrąg lub elipsa,
- $\delta = \pi$: odbicie figury przez oś,
- $\delta$ dowolne: wpływa na „skręcenie” figury, jej orientację, ale niekoniecznie na zamknięcie.

Zmiana przesunięcia fazowego daje największą wizualną różnorodność kształtów Lissajous bez zmiany częstotliwości.

## Tabela podsumowująca wpływ wszystkich parametrów

| $a$:$b$ (częstotliwości) | $A$:$B$ (amplitudy) | $\delta$ (przesunięcie fazowe) | Typ figury / cechy                                     |
|--------------------------|---------------------|----------------------------------|--------------------------------------------------------|
| 1:1                      | 1:1                 | $0$                              | Linia prosta (ukośna), $y = x$                         |
| 1:1                      | 1:1                 | $\frac{\pi}{2}$                  | Okrąg                                                  |
| 1:1                      | 2:1                 | $\frac{\pi}{2}$                  | Elipsa, rozciągnięta w poziomie                       |
| 1:1                      | 1:2                 | $\frac{\pi}{2}$                  | Elipsa, rozciągnięta w pionie                          |
| 1:1                      | 1:1                 | $\pi$                            | Linia prosta (malejąca), $y = -x$                      |
| 2:1                      | 1:1                 | $0$                              | Ósemka (figura z pętlami: 1 w pionie, 2 w poziomie)    |
| 3:2                      | 1:1                 | $\frac{\pi}{4}$                  | Zamknięta figura z 2 pętlami w pionie, 3 w poziomie    |
| 3:2                      | 2:1                 | $\frac{\pi}{2}$                  | Ta sama struktura co wyżej, ale szersza                |
| 5:3                      | 1:1                 | $0$                              | Złożona symetryczna figura (3 pionowe, 5 poziomych pętli) |
| $\sqrt{2}$:1             | 1:1                 | dowolne                          | Niezamykająca się, gęsta, „chaotyczna” struktura       |
| 1:0                      | 1:1                 | –                                | Ruch tylko w osi $x$ – linia pozioma                  |
| 0:1                      | 1:1                 | –                                | Ruch tylko w osi $y$ – linia pionowa                  |
| 1:1                      | 0:0                 | –                                | Punkt nieruchomy – brak figury                         |

> **Legenda:**
> - $a:b$ – stosunek częstotliwości w osiach $x$ i $y$.
> - $A:B$ – stosunek amplitud w osiach $x$ i $y$.
> - $\delta$ – przesunięcie fazowe między drganiami.

> **Uwagi:**
> - Przy **wymiernych $a:b$**, figury są **zamknięte**.
> - Przy **niewymiernych $a:b$**, figury są **niezamknięte** i zapełniają obszar.
> - Zmiana amplitudy nie wpływa na zamknięcie, tylko na **proporcje figury**.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Lista przypadków z tabeli (a, b, A, B, δ, opis)
cases = [
    (1, 1, 1, 1, 0, "Linia prosta (y = x)"),
    (1, 1, 1, 1, np.pi/2, "Okrąg"),
    (1, 1, 2, 1, np.pi/2, "Elipsa (rozciągnięta w poziomie)"),
    (1, 1, 1, 2, np.pi/2, "Elipsa (rozciągnięta w pionie)"),
    (1, 1, 1, 1, np.pi, "Linia prosta (y = -x)"),
    (2, 1, 1, 1, 0, "Ósemka (2:1)"),
    (3, 2, 1, 1, np.pi/4, "Zamknięta figura (3:2)"),
    (3, 2, 2, 1, np.pi/2, "Szersza figura (3:2)"),
    (5, 3, 1, 1, 0, "Złożona figura (5:3)"),
    (np.sqrt(2), 1, 1, 1, 0, "Chaotyczna struktura (√2:1)"),
    (1, 0, 1, 1, 0, "Linia pozioma (tylko oś x)"),
    (0, 1, 1, 1, 0, "Linia pionowa (tylko oś y)"),
    (1, 1, 0, 0, 0, "Punkt nieruchomy"),
]

# Parametry animacji
n_points = 1000
t_total = np.linspace(0, 24 * np.pi, n_points)

# Przygotuj dane
x_data = []
y_data = []
titles = []

for a, b, A, B, delta, title in cases:
    if a == 0:
        x = np.zeros_like(t_total)
    else:
        x = A * np.sin(a * t_total + delta)

    if b == 0:
        y = np.zeros_like(t_total)
    else:
        y = B * np.sin(b * t_total)

    x_data.append(x)
    y_data.append(y)
    titles.append(title)

# Konfiguracja wykresu
fig, axs = plt.subplots(4, 4, figsize=(20, 20))
plt.subplots_adjust(hspace=0.5, wspace=0.3)
axs = axs.flatten()

# Inicjalizacja linii
lines = []
for i, ax in enumerate(axs):
    if i < len(cases):
        x_max = max(abs(x_data[i])) if max(abs(x_data[i])) > 0 else 1
        y_max = max(abs(y_data[i])) if max(abs(y_data[i])) > 0 else 1
        ax.set_xlim(-x_max*1.1, x_max*1.1)
        ax.set_ylim(-y_max*1.1, y_max*1.1)
        ax.set_title(titles[i], fontsize=9)
        ax.grid(True, linestyle='--', alpha=0.5)
        line, = ax.plot([], [], 'b-', lw=1)
        lines.append(line)
    else:
        ax.axis('off')

# Funkcje animacji
def init():
    for line in lines:
        line.set_data([], [])
    return lines

def update(frame):
    for i in range(len(lines)):
        lines[i].set_data(x_data[i][:frame], y_data[i][:frame])
    return lines

# Stwórz animację i wyświetl JEDYNIE ją (bez wykresów statycznych)
anim = FuncAnimation(fig, update, frames=np.arange(0, n_points, 10),
                    init_func=init, blit=True, interval=30)

# Zamknij wykres po wygenerowaniu animacji, aby uniknąć podwójnego wyświetlania
plt.close(fig)

# Wyświetl tylko animację
HTML(anim.to_jshtml())
```

# Drgania harmoniczne – notatka z opisem wzorów

---

### 1. Siła sprężystości

$$
F_{\rm spręż} = -k\,x
$$

**Opis:**

- $F_{\rm spręż}$ – siła sprężyny przeciwdziałająca wychyleniu.  
- Znak minus oznacza, że działa przeciwnie do wychylenia.  
- $k$ (N/m) – sztywność sprężyny.  
- $x$ (m) – wychylenie od położenia równowagi.  

---

### 2. Równanie ruchu (Newton + Hooke)

$$
m\,\ddot{x} + k\,x = 0
$$

- $m\,\ddot{x}$ – siła bezwładności (masa × przyspieszenie).  
- $k\,x$ – siła sprężystości.  
- Układ w równowadze: suma sił = 0.  

---

### 3. Częstość kołowa

$$
\omega_0 = \sqrt{\frac{k}{m}}
$$

- $\omega_0$ (rad/s) – naturalna częstość drgań.  
- Zależy od stosunku sztywności do masy.  

---

### 4. Uproszczona postać równania

$$
\ddot{x} + \omega_0^2\,x = 0
$$

- Powstaje z podzielenia przez $m$ i podstawienia $\omega_0^2 = \tfrac{k}{m}$.  

---

### 5. Rozwiązanie – drgania sinusoidalne

$$
x(t) = A\,\sin(\omega_0\,t + \delta)
$$

- $A$ – amplituda.  
- $\omega_0\,t$ – część oscylacyjna (czasowa).  
- $\delta$ – przesunięcie fazowe.  

---

### 6. Energia potencjalna i kinetyczna

$$
E_p = \tfrac{1}{2}\,k\,x^2,
\quad
E_k = \tfrac{1}{2}\,m\,\dot{x}^2
$$

- $E_p$ – energia sprężystości, zależna od wychylenia.  
- $E_k$ – energia ruchu, zależna od prędkości.  

---

### 7. Energia całkowita układu
### Energia drgań w pojedynczym oscylatorze i w układzie dwuoscylatorów

#### Energia pojedynczego oscylatora

- **Energia potencjalna**:
  $$
  E_p = \frac{1}{2}k x^2
  $$

- **Energia kinetyczna**:
  $$
  E_k = \frac{1}{2}m \dot{x}^2
  $$

- **Energia całkowita**:
  $$
  E_{\text{tot}} = E_p + E_k = \frac{1}{2}k A^2 = \text{const.}
  $$

---

#### Energia układu dwuoscylatorów (Lissajous)

Każda oś traktujemy jako osobny oscylator.

- Parametry:

  $$
  k_x = m a^2, \quad
  k_y = m b^2
  $$

- **Energia potencjalna**:
  $$
  E_{p,x} = \frac{1}{2}k_x A^2, \quad
  E_{p,y} = \frac{1}{2}k_y B^2
  $$

- **Energia kinetyczna**:
  $$
  E_{k,x} = \frac{1}{2}m a^2 A^2, \quad
  E_{k,y} = \frac{1}{2}m b^2 B^2
  $$

- **Energia całkowita układu (bez tłumienia)**:
  $$
  E_{\text{tot}} = E_{p,x} + E_{k,x} + E_{p,y} + E_{k,y} = \frac{1}{2}m(a^2 A^2 + b^2 B^2) = \text{const.}
  $$

---

#### Energia układu tłumionego

Amplitudy maleją w czasie:

$$
A(t) = A e^{-\gamma_x t}, \quad B(t) = B e^{-\gamma_y t}
$$

- **Energia w funkcji czasu**:
  $$
  E_x(t) = \frac{1}{2}m a^2 A^2 e^{-2\gamma_x t}, \quad
  E_y(t) = \frac{1}{2}m b^2 B^2 e^{-2\gamma_y t}
  $$

- **Suma energii**:
  $$
  E_{\text{tot}}(t) = E_x(t) + E_y(t)
  $$


---

### 8. Siła tłumienia (oporu)

$$
F_{\rm opór} = -b\,\dot{x}
$$

- $b$ – współczynnik tłumienia (kg/s).  
- $\dot{x}$ – prędkość.  
- Siła przeciwdziała ruchowi.  

---

### 9. Równanie z tłumieniem

$$
m\,\ddot{x} + b\,\dot{x} + k\,x = 0
$$

- Dodano człon oporu do wcześniejszego równania.  

---

### 10. Zmienne pomocnicze

$$
\omega_0 = \sqrt{\frac{k}{m}},
\quad
\gamma = \frac{b}{2m}
$$

- $\omega_0$ – częstość własna.  
- $\gamma$ – współczynnik tłumienia (1/s).  

---

### 11. Uproszczone równanie z tłumieniem

$$
\ddot{x} + 2\gamma\,\dot{x} + \omega_0^2\,x = 0
$$

- Wygodniejsza forma równania tłumionych drgań.  

---

### 12. Rozwiązanie dla $\gamma < \omega_0$

$$
x(t) = A\,e^{-\gamma t}\,\sin(\omega\,t + \delta),
\quad
\omega = \sqrt{\omega_0^2 - \gamma^2}
$$

- $e^{-\gamma t}$ – tłumienie amplitudy.  
- $\omega$ – zmodyfikowana częstość.  

---

---

### 13. Krzywe Lissajous z tłumieniem

$$
\begin{cases}
x(t) = A\,e^{-\gamma_x t}\,\sin(\omega_x t + \delta),\\
y(t) = B\,e^{-\gamma_y t}\,\sin(\omega_y t)
\end{cases}
$$

- $A$, $B$ – początkowe amplitudy.  
- $\gamma_x$, $\gamma_y$ – tłumienie w każdej osi.  
- $\omega_x$, $\omega_y$ – częstości tłumionych drgań.  
- $\delta$ – przesunięcie fazowe.  

Efekt: punkt kreśli pętle, które z czasem się kurczą i gasną w centrum.

---

## Tabela wpływu tłumienia na ruch

|    Reżim tłumienia   |       Warunek       | Pojedynczy oscylator $x(t)$                                                                                           | Kształt tłumionych Lissajous                                                                                                                 |
| :------------------: | :-----------------: | :-------------------------------------------------------------------------------------------------------------------- | :------------------------------------------------------------------------------------------------------------------------------------------- |
|    **Underdamped**   | $\gamma < \omega_0$ | $x(t)=A\,e^{-\gamma t}\sin(\omega t+\delta)$, $\omega=\sqrt{\omega_0^2-\gamma^2}$ – oscylacje ze zanikającą amplitudą  | Pętle Lissajous stopniowo „zwijają się” do środka; proporcje zachowane, ale figura kurczy się wykładniczo                                  |
| **Critical damping** | $\gamma = \omega_0$ | $x(t)=(C_1 + C_2\,t)\,e^{-\omega_0 t}$ – najszybszy powrót do zera bez oscylacji                                      | W jednej osi ruch znika najszybciej; jeśli obie osie krytycznie tłumione – brak pętli, szybki powrót do środka                                 |
|    **Overdamped**    | $\gamma > \omega_0$ | $x(t)=C_1 e^{-(\gamma+\sqrt{\gamma^2-\omega_0^2})t}+C_2 e^{-(\gamma-\sqrt{\gamma^2-\omega_0^2})t}$ – brak oscylacji    | Składniki $x$ i $y$ „ślizgają się” wykładniczo do zera, bez tworzenia pętli – Lissajous sprowadza się do monotonicznych trajektorii          |
|  **Brak tłumienia**  |     $\gamma = 0$    | $x(t)=A\sin(\omega_0 t+\delta)$ – niegasnące oscylacje                                                                | Idealna, stacjonarna figura Lissajous (zamknięte pętle dla wymiernego $\omega_x:\omega_y$, nigdy się nie zamyka dla niewymiernego stosunku) |

- **$\gamma = \tfrac{b}{2m}$** – tempo zaniku amplitudy (1/s).  
- **$\omega_0 = \sqrt{k/m}$** – naturalna częstość drgań bez tłumienia.  
- **Underdamped** ($\gamma<\omega_0$) – widoczne oscylacje ze spadającą amplitudą.  
- **Critical damping** ($\gamma=\omega_0$) – najszybsze stłumienie bez oscylacji (np. amortyzatory).  
- **Overdamped** ($\gamma>\omega_0$) – powolny powrót bez drgań zwrotnych.  

---

### Przykład liczbowy

Dla masy $m=1\,$kg i sprężyny $k=4\,$N/m mamy $\omega_0=2\,$rad/s. Zatem:

- **Underdamped:** $b=2\,$kg/s → $\gamma=1<2$.  
- **Critical:** $b=2m\omega_0=4\,$kg/s → $\gamma=2$.  
- **Overdamped:** $b=6\,$kg/s → $\gamma=3>2$.  

Jeśli oba oscylatory w osiach $x$ i $y$ mają $\gamma_x=1$, $\gamma_y=0.5$ i częstości $\omega_{0,x}=2$, $\omega_{0,y}=3$, to pętle Lissajous będą się szybciej kurczyć w osi $x$ (większe tłumienie) niż w osi $y$.

--

---

### Wyprowadzenie czynnika wygaszającego $e^{-\gamma t}$

Aby zrozumieć, skąd dokładnie bierze się wykładniczy spadek amplitudy, prześledźmy wyprowadzenie krok po kroku:

1. **Równanie ruchu z tłumieniem**:

   $$
   m\,\ddot x + b\,\dot x + k\,x = 0.
   $$

2. **Podział przez masę i definicja $\gamma$**:

   $$
   \ddot x + \frac{b}{m}\,\dot x + \frac{k}{m}\,x = 0
   \;\Longrightarrow\;
   \ddot x + 2\gamma\,\dot x + \omega_0^2\,x = 0,
   $$

   gdzie  
   $\omega_0 = \sqrt{\tfrac{k}{m}}$ – naturalna częstość,  
   $\gamma = \tfrac{b}{2m}$ – współczynnik zaniku amplitudy.

3. **Metoda wykładniczego założenia**:  
   Zakładamy $x(t) = e^{\lambda t}$. Po podstawieniu otrzymujemy równanie charakterystyczne

   $$
   \lambda^2 + 2\gamma\,\lambda + \omega_0^2 = 0.
   $$

4. **Pierwiastki równania**:  
   Dla $\gamma < \omega_0$ rozwiązania to  
   $\lambda = -\gamma \pm i\sqrt{\omega_0^2 - \gamma^2} = -\gamma \pm i\omega$.  
   Część rzeczywista $-\gamma$ odpowiada za wykładniczy zanik.

5. **Forma końcowa**:  
   Kombinacja tych rozwiązań daje w zapisie rzeczywistym

   $$
   \boxed{x(t) = A\,e^{-\gamma t}\sin(\omega t + \delta)}.
   $$

6. **Interpretacja**:

   - $e^{-\gamma t}$ – wykładniczy czynnik wygaszający amplitudę, bezpośredni skutek tłumienia $b$.
   - $\gamma = \tfrac{b}{2m}$ – mierzy tempo utraty energii przez opór.

---


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Wybrane 4 przypadki z uwzględnieniem tłumienia i energii
cases = [
    {
        "a": 1,
        "b": 1,
        "A": 1,
        "B": 1,
        "delta": np.pi/2,
        "gamma": 0.1,  # współczynnik tłumienia
        "title": "Okrąg (a=1, b=1, δ=π/2)"
    },
    {
        "a": 2,
        "b": 1,
        "A": 1,
        "B": 1,
        "delta": 0,
        "gamma": 0.15,
        "title": "Ósemka (a=2, b=1)"
    },
    {
        "a": 3,
        "b": 2,
        "A": 1,
        "B": 1,
        "delta": np.pi/4,
        "gamma": 0.2,
        "title": "Zamknięta figura (3:2)"
    },
    {
        "a": np.sqrt(2),
        "b": 1,
        "A": 1,
        "B": 1,
        "delta": 0,
        "gamma": 0.1,
        "title": "Chaotyczna struktura (√2:1)"
    }
]

# Parametry czasu
t = np.linspace(0, 360 * np.pi, 3000)  # Dłuższy czas, aby zobaczyć tłumienie

# Przygotowanie danych
for case in cases:
    # Obliczenia x(t) i y(t) z tłumieniem
    case["x"] = case["A"] * np.exp(-case["gamma"] * t) * np.sin(case["a"] * t + case["delta"])
    case["y"] = case["B"] * np.exp(-case["gamma"] * t) * np.sin(case["b"] * t)

    # Obliczenia energii (E = 0.5 * m * (A^2 * a^2 + B^2 * b^2) * e^(-2γt))
    case["energy"] = 0.5 * (case["A"]**2 * case["a"]**2 + case["B"]**2 * case["b"]**2) * np.exp(-2 * case["gamma"] * t)

# Konfiguracja wykresów
fig, axs = plt.subplots(4, 2, figsize=(15, 20))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Inicjalizacja linii i tekstu
lines = []
energy_lines = []
for i, case in enumerate(cases):
    # Wykres krzywej Lissajous
    axs[i, 0].set_title(case["title"], fontsize=10)
    axs[i, 0].set_xlim(-1.5, 1.5)
    axs[i, 0].set_ylim(-1.5, 1.5)
    axs[i, 0].grid(True, linestyle='--', alpha=0.5)
    line, = axs[i, 0].plot([], [], 'b-', lw=1)
    lines.append(line)

    # Wykres energii
    axs[i, 1].set_title("Energia układu", fontsize=10)
    axs[i, 1].set_xlim(0, 6 * np.pi)
    axs[i, 1].set_ylim(0, 1.2 * np.max(case["energy"]))
    axs[i, 1].grid(True, linestyle='--', alpha=0.5)
    energy_line, = axs[i, 1].plot([], [], 'r-', lw=1)
    energy_lines.append(energy_line)

# Funkcja inicjalizująca
def init():
    for line in lines + energy_lines:
        line.set_data([], [])
    return lines + energy_lines

# Funkcja aktualizująca klatkę
def update(frame):
    for i, case in enumerate(cases):
        # Aktualizacja krzywej Lissajous
        lines[i].set_data(case["x"][:frame], case["y"][:frame])

        # Aktualizacja energii
        energy_lines[i].set_data(t[:frame], case["energy"][:frame])

    return lines + energy_lines

# Animacja
anim = FuncAnimation(fig, update, frames=len(t), init_func=init, blit=True, interval=35)

# Wyświetl animację
plt.close(fig)
HTML(anim.to_jshtml())
```
