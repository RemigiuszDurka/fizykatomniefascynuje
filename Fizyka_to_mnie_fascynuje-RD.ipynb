## Trochę Matematyki

### **Pochodne**

Pochodna funkcji $f(x)$ to funkcja $f'(x)$, która opisuje tempo zmian wartości funkcji $f(x)$ w zależności od zmiany argumentu $x$.

$$
\begin{array}{|c|c|}
\hline
\textbf{Funkcja } f(x) & \textbf{Pochodna } f'(x) \\ \hline
x^n & n \cdot x^{n-1} \\ \hline
\sin(x) & \cos(x) \\ \hline
\cos(x) & -\sin(x) \\ \hline
e^x & e^x \\ \hline
\ln(x) & \frac{1}{x} \\ \hline
\frac{1}{x} & -\frac{1}{x^2} \\ \hline
a^x & a^x \ln(a) \\ \hline
\tan(x) & \frac{1}{\cos^2(x)} \\ \hline
\arcsin(x) & \frac{1}{\sqrt{1 - x^2}} \\ \hline
\arccos(x) & -\frac{1}{\sqrt{1 - x^2}} \\ \hline
\arctan(x) & \frac{1}{1 + x^2} \\ \hline
\end{array}
$$

Porównanie:
$$
\begin{array}{|c|c|c|}
\hline
\textbf{Kategoria} & \textbf{Matematycy} & \textbf{Fizycy} \\ \hline
\textbf{Zmienne} & x & t \\ \hline
\textbf{Funkcje} & y(x) & x(t) \\ \hline
\textbf{Pochodna pierwsza} & \frac{dy}{dx} & \frac{dx}{dt} \quad \text{(prędkość)} \\ \hline
\textbf{Pochodna druga} & \frac{d^2y}{dx^2} & \frac{d^2x}{dt^2} \quad \text{(przyspieszenie)} \\ \hline
\textbf{Zastosowanie} & \text{Analiza matematyczna} & \text{Dynamika ciał} \\ \hline
\end{array}
$$

Naturalnie ma wiele zastosowań!

### **Równania algebraiczne i różniczkowe**

Równania ze zmienną $x$ to podstawowy rodzaj równań, w których rozwiązanie polega na znalezieniu liczby, która zapewnia tożsamościowe spełnienie równania – oznacza to, że lewa strona równa się prawej stronie dla danej wartości $x$.

Przykład równania algebraicznego:

$$
2x + 3 = 7
$$
Rozwiązaniem jest $$x = 2$$, ponieważ po podstawieniu tej wartości obie strony równania są równe:
$$
2 \cdot 2 + 3 = 7.
$$

---

### **Równania różniczkowe**

Równania różniczkowe to bardziej zaawansowany typ równań, w których zamiast szukać liczby, szukamy funkcji $x(t)$, która spełnia pewną relację różniczkową. Rozwiązanie takiego równania polega na znalezieniu funkcji, która zapewnia tożsamościowe spełnienie równania dla każdej wartości zmiennej niezależnej $t$.

Przykład równania różniczkowego:
$$
\frac{d^2 x(t)}{dt^2} = - x(t) - 2 \frac{dx(t)}{dt}.
$$

Tutaj szukamy funkcji $x(t)$, która spełnia to równanie – np. funkcja wykładnicza $x(t) = e^{-t}$ może być rozwiązaniem dla odpowiednich warunków początkowych.

Równania różniczkowe mają zastosowanie w opisie wielu zjawisk fizycznych, takich jak ruch, przenoszenie ciepła, czy dynamika populacji, i stanowią podstawę matematycznego modelowania w naukach przyrodniczych i technicznych.

## **Trochę fizyki odnośnie ruch i co na niego wpływa**


**Ruch** jest to zmiana położenia ciała w czasie. Położenie ciała określamy względem ustalonego układu odniesienia (ruch jest zawsze względem czegoś).

Prędkość to zmiana położenia w czasie. Przyspieszenie to zmiana prędkości w czasie.

### **Historia:**

- **Arystoteles**: Dużo rozmyślań na temat ruchu.
- **Galileusz**: Liczne obserwacje i eksperymenty dotyczące ruchu.
- **Newton**: Ujęcie koncepcji ruchu w sposób matematyczny.

---

### **Drugie Prawo Newtona**

Przyspieszenie obiektu jest wprost proporcjonalne do siły netto działającej na niego i odwrotnie proporcjonalne do jego masy:

$$
ma = \text{Siła}
$$

### **Isaac Newton**

*"Podajcie mi wszystkie siły działające na ciało, a przewidzę z boską/matematyczną precyzją, jak ciało się poruszy"*

$$
m \frac{d^2 x(t)}{dt^2} = \text{Siła}
$$

Rozwiązanie równania różniczkowego dla zadanej siły daje:

### **Ruch w jednym wymiarze**

Położenie to wszystko co potrzebujemy, aby opisać ruch w jednym wymiarze. Możemy je opisać jako funkcję położenia $x(t)$, prędkości $v(t)$ i przyspieszenia $a(t)$.

$$
\begin{aligned}
x(t) & & & \text{Położenie} \\
v(t) & = \frac{dx(t)}{dt} &  &\text{Prędkość} \\
a(t) & = \frac{dv(t)}{dt} = \frac{d^2 x(t)}{dt^2}  & & \text{Przyspieszenie}
\end{aligned}
$$

Tabela z paroma przykładami

$$
\begin{array}{|c|c|c|}
\hline
\textbf{Położenie } x(t) & \textbf{Prędkość } v(t) & \textbf{Przyspieszenie } a(t) \\ \hline
x_0 \text{ (stała)} & 0 & 0 \\ \hline
x_0 + vt & v \text{ (stała)} & 0 \\ \hline
x_0 + v_0t + \frac{1}{2}at^2 & v_0 + at & a \text{ (stała)} \\ \hline
A \sin(\omega t + \phi) & A \omega \cos(\omega t + \phi) & -A \omega^2 \sin(\omega t + \phi) \\ \hline
e^{kt} & ke^{kt} & k^2 e^{kt} \\ \hline
\end{array}
$$

### **Podstawy Ruchu**

- **Prędkość** wpływa na **położenie**.
- **Siła** wpływa na **prędkość**.

Podaj kod python który pokazuje rzut ukośny bez i z oporem powietrza

````python
import numpy as np
import matplotlib.pyplot as plt

# Parametry rzutu
v0 = 50  # prędkość początkowa (m/s)
angle = 45  # kąt rzutu (stopnie)
g = 9.81  # przyspieszenie grawitacyjne (m/s^2)
mass = 0.5  # masa obiektu (kg)
drag_coefficient = 0.1  # współczynnik oporu powietrza (kg/m)

# Funkcje pomocnicze
def trajectory_no_air_resistance(v0, angle, g, t):
    angle_rad = np.radians(angle)
    x = v0 * np.cos(angle_rad) * t
    y = v0 * np.sin(angle_rad) * t - 0.5 * g * t**2
    return x, y

def trajectory_with_air_resistance(v0, angle, g, mass, drag_coefficient, dt=0.01):
    angle_rad = np.radians(angle)
    vx = v0 * np.cos(angle_rad)
    vy = v0 * np.sin(angle_rad)
    x, y = 0, 0
    x_positions, y_positions = [x], [y]

    while y >= 0:
        ax = -drag_coefficient * vx / mass
        ay = -g - (drag_coefficient * vy / mass)
        vx += ax * dt
        vy += ay * dt
        x += vx * dt
        y += vy * dt
        x_positions.append(x)
        y_positions.append(y)

    return x_positions, y_positions

# Czas symulacji dla przypadku bez oporu powietrza
t_max = 2 * v0 * np.sin(np.radians(angle)) / g
t = np.linspace(0, t_max, num=500)

# Obliczenia
x_no_air, y_no_air = trajectory_no_air_resistance(v0, angle, g, t)
x_with_air, y_with_air = trajectory_with_air_resistance(v0, angle, g, mass, drag_coefficient)

# Wykres
plt.figure(figsize=(10, 6))
plt.plot(x_no_air, y_no_air, label="Bez oporu powietrza", linestyle="--")
plt.plot(x_with_air, y_with_air, label="Z oporem powietrza", linestyle="-")
plt.title("Rzut ukośny: porównanie z i bez oporu powietrza")
plt.xlabel("Odległość (m)")
plt.ylabel("Wysokość (m)")
plt.legend()
plt.grid()
plt.show()
````

Podaj kod pythona realizujący ruch ciała na sprężynie z tłumieniem.

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate

# Parametry ruchu harmonicznego tłumionego
k = 50  # stała sprężystości (N/m)
c = 1.5  # współczynnik tłumienia (kg/s)
m = 2.0  # masa (kg)
x0 = 1.0  # wychylenie początkowe (m)
v0 = 0.0  # prędkość początkowa (m/s)

# Funkcja opisująca układ równań różniczkowych
def damped_spring(t, y):
    x, v = y
    dxdt = v
    dvdt = -(k/m) * x - (c/m) * v
    return [dxdt, dvdt]

# Warunki początkowe: x, v
initial_conditions = [x0, v0]

# Czas symulacji
t = np.linspace(0, 10, 500)

# Rozwiązanie równań
solution = scipy.integrate.solve_ivp(
    damped_spring, [t[0], t[-1]], initial_conditions, t_eval=t
)

# Położenie w czasie
x = solution.y[0]

# Wykres położenia od czasu
plt.figure(figsize=(10, 6))
plt.plot(t, x, label="Położenie sprężyny z tłumieniem", linewidth=2)
plt.title("Położenie od czasu dla sprężyny z tłumieniem")
plt.xlabel("Czas (s)")
plt.ylabel("Położenie (m)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.grid(True)
plt.legend()
plt.show()
````

# Zasady Dynamiki Newtona

Isaac Newton sformułował trzy podstawowe zasady ruchu, które opisują, jak ciała poruszają się pod wpływem sił. Te zasady są fundamentem klasycznej mechaniki.

## 1. Pierwsza zasada Newtona – Zasada bezwładności
Ciało pozostaje w spoczynku lub porusza się ruchem jednostajnym prostoliniowym, dopóki nie zadziała na nie zewnętrzna siła. Oznacza to, że jeśli na ciało nie działają siły lub siły się równoważą, jego prędkość nie ulegnie zmianie.

$$ \sum \vec{F} = 0 \Rightarrow \vec{v} = \text{const} $$

Przykładem jest poruszający się samochód, który nagle hamuje – pasażerowie odczuwają siłę pchającą ich do przodu.

## 2. Druga zasada Newtona – Równanie ruchu
Przyspieszenie ciała jest proporcjonalne do działającej na nie siły i odwrotnie proporcjonalne do jego masy:

$$ \vec{F} = m \vec{a} $$

Oznacza to, że im większa siła działa na ciało, tym większe przyspieszenie uzyskuje. Na przykład, kopnięta piłka (większa siła) porusza się szybciej niż lekko popchnięta piłka.

## 3. Trzecia zasada Newtona – Zasada akcji i reakcji
Każdej akcji towarzyszy równa co do wartości i przeciwnie skierowana reakcja:

$$ \vec{F}_{AB} = -\vec{F}_{BA} $$

Jeśli ciało A działa siłą na ciało B, to ciało B działa na ciało A siłą o tej samej wartości, ale przeciwnie skierowaną. Przykładem jest odrzut rakiety – spaliny wyrzucane do tyłu powodują, że rakieta porusza się do przodu.
